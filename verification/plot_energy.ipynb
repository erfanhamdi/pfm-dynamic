```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
plt.style.use('/projectnb/lejlab2/erfan/pfm_ds/verification/theme.mplstyle')
import matplotlib.font_manager as font_manager
path = '/projectnb/lejlab2/erfan/Recode/documentation/Times_New_Roman.ttf'
prop = font_manager.FontProperties(fname=path)
mpl.rcParams['font.family'] = prop.get_name()
```

```python
data_ref = "/projectnb/lejlab2/erfan/PhaFiDyn/Results/PhaFiDyn_KalthoffWinkler_MieheAT2_2026-04-22-19h37min01sec/PhaFiDyn_KalthoffWinkler.data"
data_sim_fine = "/projectnb/lejlab2/erfan/PhaFiDyn/pfm_lite/results/tension_miehe_kalt-fixed_fine/simulation.data"
data_sim = "/projectnb/lejlab2/erfan/PhaFiDyn/pfm_lite/results/tension_miehe_kalt-fixed/simulation.data"
data_borden = "/projectnb/lejlab2/erfan/PhaFiDyn/borden_data.txt"
data_borden_frac = "/projectnb/lejlab2/erfan/PhaFiDyn/borden_data_fracture.txt"
ref_data = np.loadtxt(data_ref, delimiter = "\t")
sim_data = np.loadtxt(data_sim, delimiter = "\t")
sim_data_fine = np.loadtxt(data_sim_fine, delimiter = "\t")
borden_data = np.loadtxt(data_borden, delimiter = ",")
borden_data_frac = np.loadtxt(data_borden_frac, delimiter = ",")
plt.figure()
plt.plot(ref_data[:, 1], ref_data[:, 2], label = r"Barki et al. (2025)")
plt.plot(borden_data[:, 0]*1e-6, borden_data[:, 1], label = r"Borden et al. (2012)", linestyle = "--")
plt.plot(sim_data[:, 1], sim_data[:, 2], label = r"Ours $-$ $800 \times 800$ mesh", linestyle = "-")
plt.plot(sim_data_fine[:, 1], sim_data_fine[:, 2], label = r"Ours $-$ $1024 \times 1024$ mesh", linestyle = "-")
plt.xlabel(r"time [$\mu$s]")
plt.ylabel(r"$\mathcal{E}_{el}$ [J/m]")
plt.legend(fontsize=12)
```
